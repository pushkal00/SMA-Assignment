```python
import networkx as nx
```

```python
dolphins_graph = nx.read_gml("dolphins.gml")
print(nx.info(dolphins_graph))
print("Average path length:", nx.average_shortest_path_length(dolphins_graph))
print("Average clustering coefficient:", nx.algorithms.cluster.average_clustering(dolphins_graph))
```

```python
nx.draw(dolphins_graph)
```
